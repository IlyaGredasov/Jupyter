```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)
s = 10000
px1 = stats.norm(-1, 0.5)
x1 = np.sqrt(0.5) * np.random.randn(s) - 1
px2 = stats.norm(1, 1.5)
x2 = np.sqrt(1.5) * np.random.randn(s) + 1
```

```python
x = np.linspace(-4, 6, 1000)
plt.plot(x, px1.pdf(x))
plt.plot(x, px2.pdf(x))
```

```python
plt.hist(x1, bins=100)
plt.hist(x2, bins=100)
```
