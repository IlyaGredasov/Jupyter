```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(SUM,NON_TAX,p):
    def income(sum,percent):
        income_sum = sum*percent/100
        return 0.87*income_sum+0.13*(NON_TAX*10) if income_sum > (NON_TAX*10) else income_sum
    xs = np.linspace(0,SUM,SUM+1)
    ys = np.array([income(x,p)+income(SUM-x,p-1) for x in xs])
    plot = plt.subplot()
    plot.plot(xs,ys)
    plot.grid()
    plot.yaxis.set_major_locator(plt.MultipleLocator(2))
    print(ys.argmax(),SUM-ys.argmax())
```

```python
f(SUM=5100,NON_TAX=38,p=17)
```
