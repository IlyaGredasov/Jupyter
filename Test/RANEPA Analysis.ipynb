```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("./Свод результатов исследования (на английском).xlsx")
```

```python
df
```

```python
df = df[
    [
        "Unnamed: 0",
        "Unnamed: 1",
        "рус",
        "англ.1",
        "арабск",
        "тайск.1",
        "индонез.1",
        "китайск",
        "турецкий.1"
    ]
]
```

```python
df = df.rename(
    columns={
        "Unnamed: 0":"Вопросы",
        "Unnamed: 1":"Ответы",
        "англ.1":"англ",
        "тайск.1":"тайск",
        "индонез.1":"индонез",
        "турецкий.1":"турецкий"
    }
)
df["Итого"] = df.iloc[:, 2:].sum(axis=1)
```

```python
df
```

```python
question_indexes = df[["Вопросы"]].dropna().axes[0]
question_indexes
```

```python
question_indexes = question_indexes.to_list()
```

```python
question_indexes.append(184)
```

```python
question_indexes.remove(1)
```

```python
for i in range(len(question_indexes)-1):
    start = question_indexes[i]
    end = question_indexes[i+1]
    subdf = df[start:end]
    labels = pd.Series(subdf.Ответы).to_list()
    sizes = pd.Series(subdf.Итого).to_list()
    f = plt.figure()
    plt.pie(sizes, autopct=lambda p: "{:.1f}%".format(p))
    plt.legend(labels, loc="center left", bbox_to_anchor=(1,0.5))
    plt.suptitle(df.at[start,"Вопросы"]+"?", x=0.7)
    f.savefig(f"eng_plot{i+1}.pdf", bbox_inches='tight')
```
